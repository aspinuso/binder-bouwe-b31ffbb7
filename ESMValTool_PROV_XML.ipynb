###  Visualise a ESMValTool PROV-XML document

Prov Preamble

```python
from prov.serializers.provjson import ProvJSONSerializer
from prov.serializers.provxml import ProvXMLSerializer
from IPython.display import Image
import io
from prov.model import ProvDocument, ProvBundle, ProvException, first, Literal
from prov.dot import prov_to_dot


def provXMLtoVis(doc,format):
     
    
   # doc=ProvXMLSerializer.deserialize(ProvXMLSerializer,doc)
   
    with open(doc, "r", encoding="utf-8") as f:
        prov_doc = ProvDocument.deserialize(f, format="xml")
    dot = prov_to_dot(prov_doc)
    
    if format=="png":
        dot.write_png('PROV.png')
        return 'PROV.png' 
    if format=="svg":
        dot.write_svg('PROV.svg')
        return 'PROV.svg'
    
```
